```python
import torch
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import seaborn as sns
import numpy as np
from matplotlib import rcParams
from matplotlib import font_manager as fm
from matplotlib.colors import ListedColormap
import pandas as pd
```

## not shown in manuscript

```python
attr_mimic_dynamask = torch.load('ftscores/mimic_test_dynamask_fma.pt')
attr_mimic_extremalmask = torch.load('ftscores/mimic_test_extremalmask.pt')
```

```python
attr_mimic_dynamask.shape, attr_mimic_extremalmask.shape
```

```python
i=640
```

```python
fig, axes = plt.subplots(2, 1, figsize=(9, 3), sharex=True)

# First subplot
im0 = axes[0].imshow(attr_mimic_dynamask[i].numpy().T, cmap="cividis", aspect='auto', interpolation="nearest")
axes[0].set_xlabel("Timesteps")
axes[0].set_ylabel("Dynamask\nAttribtuions")
axes[0].set_yticks([])  # Remove y-axis labels
axes[0].grid(False)
fig.colorbar(im0, ax=axes[0], label="Value")

# Second subplot (you can change matrix and settings here)
im1 = axes[1].imshow(attr_mimic_extremalmask[i].numpy().T, cmap="cividis", aspect='auto', interpolation="nearest")
axes[1].set_xlabel("Timesteps")
axes[1].set_ylabel("ExtremalMask\nAttribtuions")
axes[1].set_yticks([])  # Remove y-axis labels
axes[1].grid(False)
fig.colorbar(im1, ax=axes[1], label="Value")

plt.tight_layout()
plt.show()
```
